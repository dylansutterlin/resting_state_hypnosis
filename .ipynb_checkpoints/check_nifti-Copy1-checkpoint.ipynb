#### Correlation of the ith, i+60th 3d image from one subject

```python
import random

single_ts = pre_data[1]
cuts = []
for i in range(nib.load(single_ts).shape[-1]):
    cuts.append(index_img(single_ts,i))

reference_images = cuts[:60]
contrast_images = cuts[60:]

correlations = []
for i in range(60):
    correlation = np.corrcoef(reference_images[i].get_fdata().flatten(), contrast_images[i].get_fdata().flatten())[0, 1]
    correlations.append(correlation)

# Print the correlations
for i, correlation in enumerate(correlations):
    print(f"Correlation between image {i+1} and image {i+61}: {correlation}")
#-------------------
# Random correlation between pairs
num_pairs = min(len(reference_images), len(contrast_images))
rand_correl = []
for _ in range(num_pairs):
    ref_image = random.choice(reference_images)
    con_image = random.choice(contrast_images)
    rand_correl = np.corrcoef(ref_image.get_fdata().flatten(), con_image.get_fdata().flatten())[0, 1]
    correlations.append(rand_correl)

# Print the correlations
for i, rand_correl in enumerate(correlations):
    print(f"Correlation between random pair {i+1}: {rand_correl}")
```

```python
from scipy import stats

# Generate a random distribution of correlations
random_correlations = np.random.choice(correlations, size=len(correlations), replace=False)

# Perform permutation test
n_permutations = 1000  # Number of permutations
permuted_t_statistics = []

for _ in range(n_permutations):
    permuted_correlations = np.random.choice(correlations, size=len(correlations), replace=False)
    t_statistic, _ = stats.ttest_1samp(permuted_correlations, popmean=0)
    permuted_t_statistics.append(t_statistic)

# Compute the observed t-statistic
observed_t_statistic, _ = stats.ttest_1samp(correlations, popmean=0)

# Compute the p-value by comparing the observed t-statistic to the permutation distribution
p_value = (np.abs(permuted_t_statistics) >= np.abs(observed_t_statistic)).mean()

# Print the test results
print("Permutation test results:")
print(f"Observed t-statistic: {observed_t_statistic}")
print(f"p-value: {p_value}")

# Check if the matched pair correlations are significantly different from random (e.g., p-value < 0.05)
if p_value < 0.05:
    print("The matched pair correlations are significantly different from random.")
else:
    print("The matched pair correlations are not significantly different from random.")
```

## 1. checking affines from 4D nifti files

```python
import nibabel as nib
from nilearn.image import index_img
import numpy as np
im = nib.load(all_files[0])
#im = index_img(im, 0)
print(im.shape, im.header.get_zooms(), im.get_fdata().max(), im.get_fdata().min(), im.affine)

def check_affines(im):
    compare = []
    for i in range(im.shape[-1]):
        aff = np.array(index_img(im, i).affine)
        compare.append(aff)
        if i > 0:
            if (aff == compare[i-1]).all() == False:
                print('Error, not same affines', i, aff, compare[i-1])
                break
def check_4D_affines(im):
    '''
    im : list of 4D images
    
    im is a list of 4D images
    check if all affines are the same
    for each subject
    '''
    compare = []
    for i in range(len(im)):
        if type(im[i]) == str:
            im[i] = nib.load(im[i])
            aff = np.array(nib.load(im[i]).affine)
        else:
            aff = np.array(im[i].affine)
        
        compare.append(aff)
        if i > 0:
            if (aff == compare[i-1]).all() == False:
                print('Error, not same affines for all subjects', i, aff, compare[i-1])
                break 
    
def check_shape(images_list):
    compare = []
    for i in range(len(images_list)):
        aff = np.array(nib.load(images_list[i]).shape)
        compare.append(aff)
        if i > 0:
            if (aff == compare[i-1]).all() == False:
                print('Error, not same shape for all subjects', i, aff, compare[i-1])
                break
    print('------------------')
    
for i, f in enumerate(all_files):
    im = nib.load(f)
    print('checking 4D TS : ', i)
    check_affines(im)
#check_4D_affines(all_files)
#check_shape(all_files)

m_im = mean_img(all_files)
print('------------------- \nMEAN IMAGE and its affine and SHAPE : ', m_im.affine, m_im.shape)
plotting.view_img(m_im, threshold=None)
```

#### when target_affine = reference_affine, the shape of the resampled image is NOT the same as the reference image. Have to specify mask_strategy =  ref_img.shape[:-1]

```python
from nilearn.maskers import NiftiMasker
from nilearn.image import resample_to_img
import nibabel as nib

ref_img = nib.load(all_files[0])
mask_strategy = 'epi'
resample_strategy = None # 'masker'

if resample_strategy == 'masker':
    # using masker to resample all images to the same shape
    voxel_masker = NiftiMasker(target_affine= ref_img.affine,target_shape = ref_img.shape[:-1], mask_strategy = mask_strategy, standardize='zscore_sample')
    voxel_masker.fit(all_files[1])
    masked_img = voxel_masker.transform(all_files[1])
    print('ref_img.shape :{}, masked image shape : {}, inverse transformed img shape {}'.format(ref_img.shape, masked_img.shape,voxel_masker.inverse_transform(masked_img).shape))  
    
else: 
    # using resample_to_img
    resampled_series = []
    for i in range(len(all_files)):
        resampled_series.append(resample_to_img(all_files[i], ref_img, interpolation='continuous'))
    
    print('ref_img.shape :{}, resampled imgages shapes {}'.format(ref_img.shape, [s.shape for s in resampled_series]))
    check_4D_affines(resampled_series)
```
